```python
from dotenv import load_dotenv
x = load_dotenv()
from crewai import Crew, Process, Agent, Task
```

```python
planejador = Agent(
    role="Planejador de viagem",
    goal="Planejador todos os detalhes de uma viagem incluindo roteiros e atividades",
    backstory="""
    VocÃª Ã© um especialista em planejamento de viagens, sempre em busca de novas aventuras e experiÃªncias.
    Seu objetivo Ã© garantir os detalhes da viagem sejam organizados de maneira eficiente e agradÃ¡vel.
    """,
    verbose=True
)
```

```python
orcamentista = Agent(
    role="OrÃ§amentista de viagem",
    goal="Estimar o custo total d euma viagem, considerando transporte, hospedagem, alimentaÃ§Ã£o e atividades",
    backstory="""
    VocÃª Ã© um analista financeiro focado em viagem. Sua missÃ£o Ã© garantir que os custos estejam dentro do
    orÃ§amento, criando estimativas precisas para cada parte da viagem.
    """,
    verbose=True
)
```

```python
planeja_roteiro = Task(
    description="Crie um roteiro detalhado para uma viagem na Europa, incluindo as cidades, atividades e transporte",
    agent=planejador,
    expected_output="""
    Um roteiro com a sequÃªncia de cidades a serem visitadas, as principais atividades e o tipo de transporte utilizado
    """
)
```

```python
estima_orcamento = Task(
    description="Calcule o orÃ§amento total da viagem com base nas cidades, transporte, hospedagem e atividades",
    agent=planejador,
    expected_output="""
    Uma estimativa detalhada de custos incluindo hospedagem, alimentaÃ§Ã£o, transporte, ingressos e demais despesas relevantes
    """
)
```

```python
viagem_crew = Crew(
    agents=[planejador, orcamentista],
    tasks=[planeja_roteiro, estima_orcamento],
    process=Process.sequential
)
```

```python
result = viagem_crew.kickoff()
```
